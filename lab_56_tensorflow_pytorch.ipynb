# Tensorflow/Keras 예제

1. 예제 1 - 선형 모델과 Gradient Descent

```python
import tensorflow as tf
import numpy as np

# 데이터 생성: y = 2x + 1 에 노이즈 추가
np.random.seed(42)
x = np.random.uniform(-1, 1, 100).astype('float32')
y = (2.0 * x + 1.0 + np.random.normal(0, 0.1, 100)).astype('float32')

# 학습할 파라미터 (w, b) 초기화
w = tf.Variable(0.0)   # 기울기
b = tf.Variable(0.0)   # 절편

learning_rate = 0.1

for epoch in range(100):
    with tf.GradientTape() as tape:
        y_pred = w * x + b          # 선형 예측
        loss   = tf.reduce_mean((y_pred - y) ** 2)  # MSE 손실

    # 손실에 대한 w, b 의 기울기 계산
    dw, db = tape.gradient(loss, [w, b])

    # 파라미터 업데이트: θ ← θ - α * ∂L/∂θ
    w.assign_sub(learning_rate * dw)
    b.assign_sub(learning_rate * db)

    if (epoch + 1) % 20 == 0:
        print(f"Epoch {epoch+1:3d} | Loss: {loss:.4f} | w={w.numpy():.3f}, b={b.numpy():.3f}")

print(f"\n정답: w=2.0, b=1.0  →  학습 결과: w={w.numpy():.3f}, b={b.numpy():.3f}")
```

```python
import torch
import numpy as np

np.random.seed(42)
x = torch.tensor(np.random.uniform(-1, 1, 100), dtype=torch.float32)
y = torch.tensor(2.0 * x.numpy() + 1.0
               + np.random.normal(0, 0.1, 100), dtype=torch.float32)

# 학습할 파라미터 (w, b) 초기화
w = torch.tensor(0.0, requires_grad=True)  # 기울기
b = torch.tensor(0.0, requires_grad=True)  # 절편

learning_rate = 0.1

for epoch in range(100):
    y_pred = w * x + b                        # 선형 예측
    loss   = ((y_pred - y) ** 2).mean()       # MSE 손실

    loss.backward()                           # 역전파: ∂loss/∂w, ∂loss/∂b 계산

    with torch.no_grad():                     # 업데이트는 미분 추적 밖에서
        w -= learning_rate * w.grad
        b -= learning_rate * b.grad

    w.grad.zero_()                            # 그래디언트 누적 방지 (필수!)
    b.grad.zero_()

    if (epoch + 1) % 20 == 0:
        print(f"Epoch {epoch+1:3d} | Loss: {loss:.4f} | w={w.item():.3f}, b={b.item():.3f}")

print(f"\n정답: w=2.0, b=1.0  →  학습 결과: w={w.item():.3f}, b={b.item():.3f}")
```
